```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('data_iris.csv')
df
```

```python
df.dtypes
```

```python
df.describe()
```

```python
plt.figure(figsize=(18,12))
plt.subplot(2,2,1)
sns.histplot(data=df, x='sepal_length',bins=10)
plt.subplot(2,2,2)
sns.histplot(data=df, x='sepal_width', bins=10)
plt.subplot(2,2,3)
sns.histplot(data=df, x='petal_length',bins=10)
plt.subplot(2,2,4)
sns.histplot(data=df,x="petal_width",bins=10)
```

```python
sns.boxplot(data=df.drop(['species'], axis=1))
```

```python
plt.figure(figsize=(18,12))
plt.subplot(2,2,1)
sns.boxplot(data=df, x='species', y='sepal_length')
plt.subplot(2,2,2)
sns.boxplot(data=df, x='species', y='sepal_width')
plt.subplot(2,2,3)
sns.boxplot(data=df, x='species', y='petal_length')
plt.subplot(2,2,4)
sns.boxplot(data=df, x='species', y='petal_width')
```

```python
plt.figure(figsize=(18,12))
plt.subplot(2,2,1)
sns.kdeplot(data=df, x="sepal_length")
plt.subplot(2,2,2)
sns.kdeplot(data=df, x="sepal_width")
plt.subplot(2,2,3)
sns.kdeplot(data=df, x="petal_length")
plt.subplot(2,2,4)
sns.kdeplot(data=df, x="petal_width")
```
